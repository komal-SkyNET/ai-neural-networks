```
import matplotlib.pyplot as plt
import gym, time
import matplotlib.animation as animation
from IPython.display import HTML
```

```
env = gym.make('SpaceInvaders-v0')
env.reset()
done = False
arr = []
fig, ax = plt.subplots()
while not done: 
  ob, rew, done, info = env.step(env.action_space.sample())
  im = ax.imshow(env.render('rgb_array'), animated=True)
  arr.append([im])
ani = animation.ArtistAnimation(fig, arr, interval=50, blit=True,
                                repeat_delay=1000)
HTML(ani.to_html5_video())
```
